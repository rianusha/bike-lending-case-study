Problem Statement : 
A bike-sharing system is a service in which bikes are made available for shared use to individuals on a short term basis for a price or free. Many bike share systems allow people to borrow a bike from a "dock" which is usually computer-controlled wherein the user enters the payment information, and the system unlocks it. This bike can then be returned to another dock belonging to the same system.

Business Goal:
To model the demand for shared bikes with the available independent variables. It will be used by the management to understand how exactly the demands vary with different features. They can accordingly manipulate the business strategy to meet the demand levels and meet the customer's expectations. Further, the model will be a good way for management to understand the demand dynamics of a new market. 

Outcomes of the analysis :
- To identify the variables affecting bike lending like seasons, temperature of the day, windspeed etc
- To create a linear model that quantitatively relates bikes lending numbers with variables such as day temperature, windspeed,holiday/weekend etc.
- To know the accuracy of the model, i.e. how well these variables can predict the number of people lending bikes.

## Step 3: Data Preparation
####  Changing certain columns into values that are categorical.

Columns like seasons, month, weekday, weathersit etc are having numerical values but are actually categorical values. So, we can replace the values with labels.


```python
season_labels = {
    1 : 'spring',
    2 : 'summer',
    3 : 'fall',
    4 : 'winter'
}

mnth_labels = {
    1 : 'January',
    2 : 'February',
    3 : 'March',
    4 : 'April',
    5 : 'May',
    6 : 'June',
    7 : 'July',
    8 : 'August',
    9 : 'September',
    10 : 'October',
    11 : 'November',
    12 : 'December'
}

weekday_labels = { 
    0 : 'Sunday',
    1 : 'Monday',
    2 : 'Tuesday',
    3 : 'Wednesday',
    4 : 'Thursday',
    5 : 'Friday',
    6 : 'Saturday'
}

weathersit_labels = {
    1 : 'Clear/Cloudy Sky',
    2 : 'Misty Sky',
    3 : 'Light snow/rain/thunderstorm'
}

# replacing numerals with labels 
bike_lending['season'] = bike_lending['season'].replace(season_labels)
bike_lending['mnth'] = bike_lending['mnth'].replace(mnth_labels)
bike_lending['weekday'] = bike_lending['weekday'].replace(weekday_labels)
bike_lending['weathersit'] = bike_lending['weathersit'].replace(weathersit_labels)

bike_lending.head()
```

####  Dummy Variables

The variables  'mnth', 'weekday', 'season' and 'weathersit' all have fixed set of values, hence we can create dummy variables for all these 4.

```python
season_dummy  = pd.get_dummies(bike_lending['season'],drop_first=True)
weather_dummy = pd.get_dummies(bike_lending['weathersit'],drop_first=True)
weekday_dummy = pd.get_dummies(bike_lending['weekday'],drop_first=True)
month_dummy = pd.get_dummies(bike_lending['mnth'],drop_first=True)
```

```python
bike_lending = pd.concat([bike_lending,season_dummy],axis=1)
bike_lending = pd.concat([bike_lending,weather_dummy],axis=1)
bike_lending = pd.concat([bike_lending,weekday_dummy],axis=1)
bike_lending = pd.concat([bike_lending,month_dummy],axis=1)
bike_lending.head()
```

We can now remove the columns 'mnth', 'weekday', 'season' and 'weathersit' as there are dummies created for them

```python
bike_lending.drop(['season','weathersit','weekday','mnth'], axis = 1, inplace = True)
```

```python
bike_lending.info()
```

## Step 4: Splitting the Data into Training and Testing Sets
##### First basic step for regression is performing a train-test split

##### Import required libraries  and split data

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(bike_lending, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python

from sklearn.preprocessing import MinMaxScaler 
numerical_scaler = MinMaxScaler()
numerical_variables = ['temp','hum','windspeed']

df_train[numerical_variables] = numerical_scaler.fit_transform(df_train[numerical_variables])
```

Lets create the X and Y variables for model building

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

## Step 5: Building a linear model 

```python
#Build a linear model
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_l = sm.OLS(y_train, X_train_lm).fit()
lr_l.params

print(lr_l.summary())
```

As we can see, there are lot of variables that have a very high p-value and aren't really significant (in the presence of other variables).So, we might need to look at feature elimination and drop some features.

 First step for building a model is to select the features set for the model. We can either manually check for each variable and check for the p and R2 values to decide on whether the variable makes sense. And keep adding more vaiables. Or else we can also use the RFE (Recursive Feature elimination) method to choose the top x variables. 
 
 In our case , lets use a hybrid approach.
 - Use RFE and pick the top 15 variables

- And then use p-values to remove any more variables if needed.

```python
# imprt sklearn library
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lr_estimator = LinearRegression()
rfe = RFE(lr_estimator,n_features_to_select=15, step=1)
selector = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col_selected =  X_train.columns[rfe.support_]
col_selected
```

```python
## Create the X now with only the rfe columns
X_train_rfe = X_train[col_selected]
X_train_rfe
```

```python
X_train_rfe_lm1 = sm.add_constant(X_train_rfe)

lr_1 = sm.OLS(y_train, X_train_rfe_lm1).fit()
lr_1.params
```

```python
print(lr_1.summary())
```

The variable January has a high p-value. So, lets eleminate the variable.

```python
X_train_rfe = X_train_rfe.drop(['January'],axis=1)
```

```python
# Model 2
X_train_rfe_lm2 = sm.add_constant(X_train_rfe)

lr_2 = sm.OLS(y_train, X_train_rfe_lm2).fit()
lr_2.params
```

```python
print(lr_2.summary())
```

```python
# Calculate the VIFs again for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Removing hum from the data set due to its high VIF value

```python
X_train_rfe = X_train_rfe.drop(['hum'],axis=1)
```

```python
X_train_rfe.head()
```

```python
#Model 3
X_train_rfe_lm3 = sm.add_constant(X_train_rfe)

lr_3 = sm.OLS(y_train, X_train_rfe_lm3).fit()
lr_3.params
```

```python
print(lr_3.summary())
```

```python
# Calculate the VIFs again for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Dropping column workingday due to its high VIF, we keep temp still as a feature due to the fact it makes sense in having that in model from a business perspective.


```python
X_train_rfe = X_train_rfe.drop(['workingday'],axis=1)
```

```python
X_train_rfe.head()
```

```python
#Model 4
X_train_rfe_lm4 = sm.add_constant(X_train_rfe)

lr_4 = sm.OLS(y_train, X_train_rfe_lm4).fit()
lr_4.params
```

```python
print(lr_4.summary())
```

```python
# Calculate the VIFs again for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Dropping column Saturday because of high p-value

```python
X_train_rfe = X_train_rfe.drop(['Saturday'],axis=1)
```

```python
# Model 5
X_train_rfe_lm5 = sm.add_constant(X_train_rfe)

lr_5 = sm.OLS(y_train, X_train_rfe_lm5).fit()
lr_5.params
```

```python
print(lr_5.summary())
```

```python
# Calculate the VIFs again for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Now as you can see, the VIFs and p-values both are within an acceptable range. So we go ahead and make our predictions using this model only.
